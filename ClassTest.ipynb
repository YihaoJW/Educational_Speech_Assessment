```python
from AsrLcss import AsrLcss
from glob import glob
from nltk.stem import LancasterStemmer
```

```python
# Load the data
data = glob("Result/*.plist")
```

```python
LCSS_CAL = AsrLcss(dict_path='yrs12_passages.csv',stemmer=LancasterStemmer())
```

```python
it = iter(data)
```

```python
while True:
    try:
        path = next(it)
        LCSS_CAL.read_student_text(path)
        break
    except KeyError:
        continue
    except StopIteration:
        break
```
