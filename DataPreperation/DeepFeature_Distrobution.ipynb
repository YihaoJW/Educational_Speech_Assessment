```python
import numpy as np
from pathlib import Path
```

```python
dirs = Path("/Users/jackwang/Documents/PhD_Research/Prof_Larson/ASR/Code/dsProject/DataFolder/Siri_Related/Deep_Feature_model_9_Numpy/")
```

```python
npy_files = sorted(dirs.glob("*.npy"))

# lists to store E[X], E[X^2], and the number of data points for each file
means = []
squares = []
sizes = []

for file in npy_files:
    data = np.load(file)
    data = data.flatten()
    data = data[data > -800]
    sizes.append(data.size)
    means.append(np.mean(data))
    squares.append(np.mean(data**2))

# converting lists to numpy arrays
sizes = np.asarray(sizes)
means = np.asarray(means)
squares = np.asarray(squares)

# calculate weighted means and variances
mean = np.average(means, weights=sizes)
variance = np.average(squares, weights=sizes) - mean**2

print(f'Overall mean: {mean}')
print(f'Overall variance: {variance}')
```
